```python
import pandas as pd
import os
```

```python
if not os.path.exists('books.csv'):
    print("File 'books.csv' does not exist.")
if not os.path.exists('ratings.csv'):
    print("File 'ratings.csv' does not exist.")

# Load the datasets
books = pd.read_csv("books.csv", sep=';', encoding="ISO-8859-1", on_bad_lines="warn")
ratings = pd.read_csv("ratings.csv", sep=";", encoding="ISO-8859-1", on_bad_lines="warn")
```

```python
# Convert 'Year-Of-Publication' to numeric, coercing errors to NaN so non-numeric values (like publisher names) become NaN
books["Year-Of-Publication"] = pd.to_numeric(books["Year-Of-Publication"], errors="coerce")
books["Year-Of-Publication"] = books["Year-Of-Publication"].fillna(books["Year-Of-Publication"].mode()[0]).astype(int)
books["Year-Of-Publication"] = books["Year-Of-Publication"].fillna(books["Year-Of-Publication"].mean()).astype(int)
```

```python
ratings["Book-Rating"] = pd.to_numeric(ratings["Book-Rating"])
ratings["Book-Rating"] = ratings["Book-Rating"].fillna(ratings["Book-Rating"].mean())
```

```python
print("Books dataset:")
print(books.info())
print("\n\nRatings dataset:")
print(ratings.info())
```

```python
df = pd.merge(books, ratings, left_on="ISBN", right_on="ISBN", how="inner")

df.head(10)
```

```python
df.isnull().sum()
```

```python
df2 = df.drop(columns=["Image-URL-L"])
```

```python
df2.columns
```

```python
aggregated_df = df2.groupby(['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher',
       'Image-URL-S', 'Image-URL-M']).agg({"User-ID":"count", "Book-Rating":"mean"}).reset_index().rename(columns={"User-ID":"Number", "Book-Rating":"Avg_Rating"})
```

```python
new_columns = {
    "ISBN": "isbn",
    "Book-Title": "title",
    "Book-Author": "author",
    "Year-Of-Publication": "year",
    "Publisher": "publisher",
    "Image-URL-S": "image_url_s",
    "Image-URL-M": "image_url_m",
    "Avg_Rating": "ratings",
}

aggregated_df.rename(columns=new_columns, inplace=True)
aggregated_df["ratings"] = aggregated_df["ratings"].round(2)    
```
